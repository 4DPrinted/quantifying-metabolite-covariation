**Import necessary packages**

```python
import pandas as pd
```

**Sequence and process metadata**

```python
meta_data = pd.read_csv("../Data/original_meta_data.txt", sep = "\t").set_index("CCLE_ID")
meta_data= meta_data.transpose()
media_components = pd.read_csv("../Data/Generated Data/media.csv").set_index("Cell_culture_media")
media_mappings = pd.read_csv("../Data/Generated Data/media.cells.csv").set_index("Cell_Lines")
media_mappings = media_mappings.dropna(axis = 0, subset = ["Cell_culture_media"])
media_mappings = media_mappings.replace({"IMDM":"DMEM:IMDM=1:1"})    
for cell_line in meta_data.columns:
    if cell_line in media_mappings.index:
        meta_data.loc["Growth.Medium"][cell_line] = media_mappings.loc[cell_line]["Cell_culture_media"]
    else:
        meta_data.loc["Growth.Medium"][cell_line]  = None
for component in media_components.columns:
    meta_data.loc[component] = ''
    for cell_line in meta_data.columns:
        if cell_line in media_mappings.index:
            print(cell_line)
            meta_data.loc[component][cell_line] = media_components.loc[media_mappings.loc[cell_line]["Cell_culture_media"]][component]
        else:
            meta_data.loc[component][cell_line] = None
```

**Write out metadata**

```python
meta_data.to_csv("../Data/Generated Data/meta_data.csv", index_label = False)
```
